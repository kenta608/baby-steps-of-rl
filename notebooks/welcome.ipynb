```python
import numpy as np
from tensorflow.python import keras as K
import gym
import gym_ple

import matplotlib.pyplot as plt
%matplotlib inline
from IPython import display

def welcome():
    """
    Code to check installation of basic libraries
    """

    env = gym.make("Catcher-v0")
    num_action = env.action_space.n
    action_dic = {0: "neutral", 1: "right", 2: "left"}
    # episode_count = 10
    episode_count = 1

    s = env.reset()
    brain = K.Sequential()
    brain.add(K.layers.Dense(num_action, input_shape=[np.prod(s.shape)],
                             activation="softmax"))

    def policy(s):
        evaluation = brain.predict(np.array([s.flatten()]))
        return np.argmax(evaluation)

    for e in range(episode_count):
        s = env.reset()
        done = False
        while not done:
            # env.render(mode="human")
            plt.imshow(env.render(mode="rgb_array"))
            display.display(plt.gcf())
            display.clear_output(wait=True)
            a = policy(s)
            n_state, reward, done, info = env.step(a)
            s = n_state
            plt.title(f"episode={e}, reward={reward}, action={action_dic[a]}, done={done}")
```

```python
welcome()
```
